```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.amazon.in/gp/bestsellers/electronics/ref=zg_bs_pg_2?ie=UTF8&pg=1"
page=requests.get(url)
soup=BeautifulSoup(page.content,'lxml')
soup
```

```python
all_data=soup.find_all(class_="zg-item-immersion")
len(all_data)
```

```python
first=all_data[0]
fm=first.find(class_="p13n-sc-truncate").get_text().strip()
fm
```

```python
rate=first.find('i',{"class":"a-icon a-icon-star a-star-4 aok-align-top"}).get_text()
rate
```

```python
rev=first.find(class_="a-size-small a-link-normal").get_text().replace(',','')
rev
```

```python
price=first.find(class_="p13n-sc-price").get_text()
price
```
